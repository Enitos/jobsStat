Total de oferta do mês de Outubro

```python

out= df.loc[(df['day'] >= '2023-10-01') & (df['day'] <= '2023-10-31')]
out_titulo=out['title'].count()
out_titulo
```

Total de oferta por mês e por posto (Outubro)

```python

porPoint_out =out['number_points'].sum()#soma  nª de posto
total_mes_out= out['number_points'].count()#conta nª de linhas na coluna number_points 

print('Por post:', porPoint_out)
print('total mes:', total_mes_out)
```

Total de oferta do mês de Novembro

```python

nov= df.loc[(df['day'] >= '2023-11-01') & (df['day'] <= '2023-11-30')]
nov_titulo= nov['title'].count()
nov_titulo
```

Total de oferta por mês e por posto (Novembro)

```python

porPoint_nov =nov['number_points'].sum()#soma  nº de posto
total_mes_nov= nov['number_points'].count()#conta nº de linhas na coluna number_points 

print('Por post:', porPoint_nov)
print('total mes:', total_mes_nov)
```

Diferença de nº de ofertas entre Outubro e Novembro

```python

diferenca_oft= total_mes_nov - total_mes_out
diferenca_oft
```

Diferença de nº de posto entre Outubro e Novembro

```python

diferenca_post= porPoint_nov - porPoint_out
diferenca_post
```

```python

import pandas as pd
import plotly.express as px
from collections import Counter
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
```

```python
# Carregar os dados do arquivo CSV
url = "https://raw.githubusercontent.com/Enitos/omt-job/main/out.csv"
data = pd.read_csv(url)

# Criar a aplicação Dash
app = JupyterDash(__name__)

# Calcular a contagem dos valores da coluna 'title'
title_counts = dict(Counter(data['title']))

# Layout do dashboard
app.layout = html.Div([
    html.H1("Vagas por cidade"),
    dcc.Graph(
        id='bar-chart',
        figure=px.bar(
            x=list(title_counts.keys()),
            y=list(title_counts.values()),
            labels={'x': 'Título da Vaga', 'y': 'Número de Ocorrências'},
            title='Número de Vagas por Título'
        )
    )
])

# Executar o aplicativo no Jupyter Notebook
app.run_server(mode='inline')
```
